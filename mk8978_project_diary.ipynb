```python
kmeans10 = cluster.KMeans(n_clusters=10)
```

```python
clusteringData_df = dfmergeb[['EmploymentRank', 'Percentage never worked']]
clusteringData_df.head()
```

```python
# Call the fit function on the clusterer to fit the data to the 10 means.
assignedClusters_clust = kmeans10.fit(clusteringData_df)
```
